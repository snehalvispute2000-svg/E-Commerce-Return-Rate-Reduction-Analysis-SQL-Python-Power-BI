## Missing Value Analysis

The request_date column contains 32,597 missing values.

After validation, these missing values were found only for orders where returned = 'No'.

Since customers who do not return an order do not have a return request date, these missing values are expected and were retained. No imputation was performed because they accurately represent the business process.

```python
# Remove Extra Spaces

text_columns = df.select_dtypes(include=['object']).columns

for col in text_columns:
  df[col] =df[col].str.strip()
```

```python
# Standardize Text

df['customer_gender'] = df['customer_gender'].str.title()

df['payment_method'] = df['payment_method'].str.title()

df['category'] = df['category'].str.title()

df['region'] = df['region'].str.title()

df['returned'] = df['returned'].str.title()
```

```python
# Negative Values

df[df['price'] < 0]

df[df['quantity'] <= 0]

df[df['shipping_cost'] < 0]
```

```python
# Date Validation

(df['delivered_date'] < df['order_date']).sum()
```

```python
(df['request_date'] < df['order_date']).sum()
```

```python
# Export Clean Dataset

df.to_csv("Ecommerce_Clean.csv", index=False)
```
